```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline


import seaborn as sns
```

```python
dataset=pd.read_excel('location.xlsx')
```

```python
dataset.head()
```

```python
dataset.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
dataset.iloc[:,:62].values
```

```python
#Using randomforest to drop some unnecessary location.(label=location)
from sklearn.ensemble import RandomForestRegressor,RandomForestClassifier
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import mean_squared_error as MSE
from sklearn.metrics import accuracy_score,r2_score
from xgboost import XGBRegressor
```

```python
X_train,X_test,y_train,y_test=train_test_split(dataset.iloc[:,:62].values,dataset['location'],test_size=0.3,random_state=42)
```

```python
param={'learning_rate':np.arange(0.1,0.5,0.1),'max_depth':np.arange(0,6,1),'min_child_weight':np.arange(0,6,1)}
```

```python
grid=GridSearchCV(estimator=XGBRegressor(),param_grid=param,cv=5)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
xgb= XGBRegressor(learning_rate=grid.best_params_['learning_rate'],max_depth=grid.best_params_['max_depth'],min_child_weight=grid.best_params_['min_child_weight'])#{'learning_rate': 0.03, 'max_depth': 4, 'min_child_weight': 4}
```

```python
xgb.fit(X_train,y_train)
```

```python
y_pred1=xgb.predict(X_test)
```

```python
rmse=MSE(y_test,y_pred1)**1/2
rmse
```

```python
y_pred=xgb.predict(X_train)
```

```python
rmse=MSE(y_train,y_pred)**1/2
rmse
```

```python
r2_score(y_train,y_pred)
```

```python
r2_score(y_test,y_pred1)
```
